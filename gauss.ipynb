```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
from scipy.stats import norm
```

```python
def gaussian(x,mu,sigma):
  return 1/(sigma*np.sqrt(2*np.pi))*np.exp(-1/2*(pow(((x-mu)/sigma),2)))
```

```python
x = np.arange(-4,4,0.1)
y = gaussian(x, 0, 1)

plt.plot(x, y)
```

```python
dist = norm(0,1)

x = np.arange(-4,4,0.1)
y = [dist.pdf(value) for value in x]
#for value in x:
  #y.append(dist.pdf(value))
plt.plot(x,y)
```

```python
dist = norm(0,1)

x = np.arange(-4,4,0.1)
y = [dist.cdf(value) for value in x]
#for value in x:
  #y.append(dist.pdf(value))
plt.plot(x,y)
```

```python
df = pd.read_excel('s057.xls')

arr = df['Normally Distributed Housefly Wing Lengths'].values[3:]

values, dist = np.unique(arr, return_counts=True)

plt.bar(x=values, height=dist)
```

```python
mu = arr.mean()
sigma = arr.std()

dist = norm(mu,sigma)
x = np.arange(30,60, 0.1)
y = [dist.pdf(value) for value in x]

values, dist = np.unique(arr,return_counts=True)
plt.plot(x,y)

plt.bar(x=values, height=dist/len(arr))
```
