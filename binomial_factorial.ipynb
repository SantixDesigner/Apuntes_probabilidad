```python
import numpy as np
from numpy.random import binomial
from scipy.stats import binom
from math import factorial
import matplotlib.pyplot as plt
```

```python
def f(k,n,p):
  return (factorial(n)/(factorial(k)*factorial(n-k))*pow(p, k)*pow(1-p, n-k))
```

```python
f(2,3,0.5)
```

```python
dist = binom(3,0.5)
dist.cdf(3)
```

```python
np.round(dist.pmf(2),3)
```

```python
def hist_plot(num_trials, n, p):
  values = [0,1,2,3]
  arr = []
  for i in range(0, num_trials):
    arr.append(binomial(n,p))
  distribucion_simulada = np.unique(arr, return_counts=True)[1]/len(arr)
  distribucion_teorica = [binom(n,p).pmf(k) for k in values ]
  plt.bar(x=values, height=distribucion_teorica, label = 'teoría', color = 'red')
  plt.bar(x=values, height=distribucion_simulada)
  plt.title('simulación con {} experimentos'.format(num_trials))

hist_plot(50, 3, 0.5)
```
